#1-Liner Interpolation

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#Interpolate the data using liner imterpolation
x_new = np.linspace(1,5,10)
y_interp= np.interp(x_new,x,y)
print(y_interp)
```

```python
plt.scatter(x_new,y_interp)
```

##Cubic interpolation with scipy.

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([2,8,27,64,1125])
```

```python
from scipy.interpolate import interp1d
```

```python
#Create cubic interpolation function
f = interp1d(x,y ,kind= 'cubic')
```

```python
#Interpolate the data using cubic imterpolation
x_new = np.linspace(1,5,10)
y_interp= f(x_new)
print(y_interp)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```

#Polynonial interpolation

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
```

```python
#Interpolate the data using interpolation.
p = np.polyfit(x,y,4) #fit a 2nd degree in polynomial data.
```

```python
x_new = np.linspace(1,5,10) #Create a new x values
y_interp = np.polyval(p,x_new) #interpolate y values
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```
